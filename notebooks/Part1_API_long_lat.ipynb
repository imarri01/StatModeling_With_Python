```python
import requests
import pandas as pd

# Make the request to the API
response = requests.get('http://api.citybik.es/v2/networks/sobi-hamilton')

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Parse the JSON response

    # Extract station information
    stations = data['network']['stations']

    # list of dictionaries with longitude and latitude for each station
    station_locations = [{'available_bikes': station['free_bikes'], 'longitude': station['longitude'], 'latitude': station['latitude']} for station in stations]

    # Create a pandas DataFrame
    df = pd.DataFrame(station_locations)

    print(df)
else:
    print("Request failed with status code:", response.status_code)
```
